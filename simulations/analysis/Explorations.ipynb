```r
count_bins <- function(x, n) {
    bins <- table(cut(x, n))
    d <- data.table(bins)
    colnames(d) <- c('interval','frequency')
    d
}
```

```r
count_bins(allele_freqs[seed == 5196]$start, 100)
```

```r
get_unique_sites_frequency_range <- function(lower, upper, freqs) {
    # https://stackoverflow.com/q/16289182
    freqs[
        freqs[
            start > lower & start < upper,
            .I[sample(.N, 1, replace=FALSE)],
            # by=c("start", "seed", "deme")
            by=c("start", "seed")
        ]$V1
    ]
}
```

```r
# f <- get_unique_sites_frequency_range(0.01, 0.05, filtered_allele_freqs)
f <- get_unique_sites_frequency_range(0.05, 0.13, filtered_allele_freqs)
f <- f[mean_vg, on='seed']
f[, mean_vg_bin := cut(f$mean_vg, 5, include.lowest = TRUE)]
f[, start_bin := cut(f$start, 5, include.lowest = TRUE)]
```
